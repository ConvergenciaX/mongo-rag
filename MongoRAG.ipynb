```python
from sentence_transformers import SentenceTransformer

# Load the embedding model (https://huggingface.co/nomic-ai/nomic-embed-text-v1")
model = SentenceTransformer("nomic-ai/nomic-embed-text-v1", trust_remote_code=True)
    
# Define a function to generate embeddings
def get_embedding(data):
    """Generates vector embeddings for the given data."""

    embedding = model.encode(data)
    return embedding.tolist()
```

```python
pdf_path1 = "C:/Users/joelc/OneDrive - Universidad Católica de Chile/GovTech-PermisosEdificacion/LEY-20071_22-NOV-2005.pdf"
pdf_path2 = "C:/Users/joelc/OneDrive - Universidad Católica de Chile/GovTech-PermisosEdificacion/Ley-20703_05-NOV-2013.pdf"
chunk_size=400
chunk_overlap=20
#pdf_path = "ruta/al/archivo.pdf"
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load the PDF
loader = PyPDFLoader(pdf_path1)
data = loader.load()
# Split the data into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=20)
documents = text_splitter.split_documents(data)
print(documents)
print("\n\n ")

loader2 = PyPDFLoader(pdf_path2)
data2 = loader2.load()
# Split the data into chunks

documents2 = text_splitter.split_documents(data2)

print(documents2)


```
